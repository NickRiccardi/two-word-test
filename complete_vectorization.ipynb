```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
```

```python
#import Graves word pair list and clean word pair format to match incoming vector names

graves_og = pd.read_excel('C:/Users/Nick/cleaned_words.xlsx')

clean_pairs = pd.concat([graves_og['Stim'], graves_og['Stim'].str.split(' ', expand=True)], axis=1)
clean_pairs = clean_pairs.drop(columns=0)
clean_pairs[2] = clean_pairs[2].str[:-1]
clean_pairs = clean_pairs.rename(columns={1:'Word1', 2:'Word2'})
word1 = pd.DataFrame(clean_pairs['Word1'])
word2 = pd.DataFrame(clean_pairs['Word2'])

graves_og['pairs'] = clean_pairs['Word1'] + ' ' + clean_pairs['Word2']

# print(clean_pairs)

print(graves_og)
```

```python
w2v = pd.read_csv('C:/Users/Nick/W2V.csv')
glove = pd.read_csv('C:/Users/Nick/glove.csv')
tax = pd.read_csv('C:/Users/Nick/tax.csv')
```

```python
word1_vec = word1.merge(w2v, how='left', left_on='Word1', right_on='Word').drop(columns='Word')
word2_vec = word2.merge(w2v, how='left', left_on='Word2', right_on='Word').drop(columns='Word')

word1_glo = word1.merge(glove, how='left', left_on='Word1', right_on='Word').drop(columns='Word')
word2_glo = word2.merge(glove, how='left', left_on='Word2', right_on='Word').drop(columns='Word')

word1_tax = word1.merge(tax, how='left', left_on='Word1', right_on='Word').drop(columns='Word')
word2_tax = word2.merge(tax, how='left', left_on='Word2', right_on='Word').drop(columns='Word')

print(word1_tax)
print(word2_tax)
```

```python
#clean_pairs['W2V_sim'] = np.dot(A,B)/(norm(A)*norm(B))
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

word_pair = []
w2v_cos = []
glo_cos = []
tax_cos = []

for i in range(len(word1_vec.index)):
    
    #first do w2v
    a=np.array(word1_vec.iloc[i,1:]).reshape(1, -1)
    b=np.array(word2_vec.iloc[i,1:]).reshape(1, -1)
    w1 = word1_vec.iloc[i,0]
    w2 = word2_vec.iloc[i,0]
    
    cos_sim = cosine_similarity(a,b)
    word_pair.append(w1 + ' ' + w2)
    w2v_cos.append(float(cos_sim))
    
    #now do glo
    a=np.array(word1_glo.iloc[i,1:]).reshape(1, -1)
    b=np.array(word2_glo.iloc[i,1:]).reshape(1, -1)
    cos_sim = cosine_similarity(a,b)
    glo_cos.append(float(cos_sim))
    
    #now do tax
    a=np.array(word1_tax.iloc[i,1:]).reshape(1, -1)
    b=np.array(word2_tax.iloc[i,1:]).reshape(1, -1)
    cos_sim = cosine_similarity(a,b)
    tax_cos.append(float(cos_sim))
```

```python
words_and_cos = {'pairs': word_pair, 'Word2Vec': w2v_cos, 'GloVe': glo_cos, 'Taxonomic': tax_cos}
words_with_sims = pd.DataFrame(words_and_cos)
print(words_with_sims)
```

```python
graves_with_vecs = graves_og.merge(words_with_sims, how='left', left_on='pairs', right_on='pairs')
print(graves_with_vecs)
graves_with_vecs.to_csv('graves_with_vecs.csv')
```
