```python
import pandas as pd
```

```python
#this gets representation vectors for the single words from Graves using SCOPE - WORK IN PROGRESS
#Right now uses Word2Vec, GloVE, and Taxonomic

#Still need to learn how to calculate representational similarity for the word pairs

#I'm printing the output dataframes to excel sheets so that I just have to read those in instead of 'all_data'

#so much data, takes a long time

all_data = pd.read_excel('C:/Users/Nick/Downloads/data_with_metadata.xlsx')
```

```python
all_data.head(10)
```

```python
#only keep the words and the vector of your choosing (here is Word2Vec)

word_vec = all_data[['Word', 'Word2Vec']]
word_vec = word_vec.dropna()

#split since all 300 vectors were in a single cell

word_vec = pd.concat([word_vec['Word'], word_vec['Word2Vec'].str.split('|', expand=True)], axis=1)
word_vec.head(20)
```

```python
single_words = pd.read_csv('C:/Users/Nick/Downloads/GPT-for-SCOPE.csv')

#this replace dict was after noticing the key in this dict were how it was listed in word2vec, while the value 
#is how it was listed in the Graves

#I replaced them originally manually - need to find better way

replace_dict = {'Bible': 'bible', 'tight': 'tights', 'ax': 'axe', 'harbor': 'harbour'}

print(single_words)
```

```python
single_with_ratings = single_words.merge(word_vec, how='left', left_on='Single Word', right_on='Word')
single_with_ratings = single_with_ratings.dropna().drop_duplicates().replace(to_replace=replace_dict).drop(columns='Single Word')
print(single_with_ratings)
```

```python
#only keep the words and the vector of your choosing (here is Word2Vec)

glove = all_data[['Word', 'GloVe']]
glove = glove.dropna()

#split since all 300 vectors were in a single cell

glove = pd.concat([glove['Word'], glove['GloVe'].str.split('|', expand=True)], axis=1)
glove.head(20)
```

```python
#check this - there might be too many unique instances? Word2vec only has 396, this has 399

single_with_glove = single_words.merge(glove, how='left', left_on='Single Word', right_on='Word')
single_with_glove = single_with_glove.dropna().drop_duplicates().replace(to_replace=replace_dict).drop(columns='Single Word')
print(single_with_glove)
```

```python
#only keep the words and the vector of your choosing (here is Word2Vec)

tax = all_data[['Word', 'Taxonomic']]
tax = tax.dropna()

#split since all 300 vectors were in a single cell

tax = pd.concat([tax['Word'], tax['Taxonomic'].str.split('|', expand=True)], axis=1)
tax.head(20)
```

```python
#maybe word2vec is the one missing words? This has 399 just like GLOVE

single_with_tax = single_words.merge(tax, how='left', left_on='Single Word', right_on='Word')
single_with_tax = single_with_tax.dropna().drop_duplicates().replace(to_replace=replace_dict).drop(columns='Single Word')
print(single_with_tax)
```

```python
single_with_ratings.to_csv('W2V.csv', index=False)
single_with_glove.to_csv('glove.csv', index=False)
single_with_tax.to_csv('tax.csv', index=False)
```
